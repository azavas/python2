### 1.1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kc_house_data.csv')
df.head()
```

### Задание 1.2

```python
plt.figure(figsize = (6,4))
plt.hist(df['price'], bins=100, color='#AA0000')
plt.grid()
plt.title('Стоимость недвижимости')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

*Вывод: Большая часть домов продаются по цене до 2 млн.рублей, но максимальное количество приходится в ценовой диапозон от 400 до 700 тыс.руб*

### Задание 1.3

```python
plt.figure(figsize = (6,4))
sns.histplot(df['sqft_living'], bins = 100, color='#FFFF00')
plt.grid()
plt.title('Распределение жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество');
```

*Максимальное количество домов продаются с площадюь от 1600 до 2300 квдратных метров*

### Задание 1.4

```python
data = df['yr_built'].value_counts()
data
```

```python
names = data.index
values = data.values
```

```python
plt.figure(figsize=(10, 6))
plt.bar(names, values)
plt.title('Годы постройки домов')
plt.xlabel('Годы')
plt.ylabel('Количество домов')
plt.xticks(rotation=50);
```

*Максимальное количество домов было построено в 2000 годах, также был прирост строительства в 60 - 70 годах*

### Задание 2.2

### Задание 2.3

```python
data_condition = df['condition'].value_counts()
data_condition
```

```python
import numpy as np
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(11, 10))
upp_mat = np.triu(df.corr())
sns.heatmap(corr_matrix, annot=True, linewidths=.5,
            cmap='coolwarm', mask= upp_mat);
```
